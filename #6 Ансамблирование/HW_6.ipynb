```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from jupyterthemes import jtplot
from sklearn.metrics import auc, roc_curve, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
jtplot.style('onedork')
```

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
data.info()
```

```python
# Удалю колонки в которых много пропусков
data.drop(['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature', 'LotFrontage'], axis=1, inplace=True)
```

```python
data.info()
```

```python
# Тк много колонок в которых есть по несколько пропусков и общая сумма не очень большая, воспользуюсь dropna
data = data.dropna()
```

```python
data.describe()
```

```python
# Будем считать, что выбросов нет, хотя есть несколько вещей, которые странные и надо бы проверить, но не суть
```

```python
# # Тк датасет очень маленький, а признаком очень много, и большинство из них категориальные, 
# и задача не стоит в максимально качественной модели, то для простоты работы, 
# я отделю часть категориальных признаков меньше 5 и далее сделаю onehotencoding 
```

```python
cat_feat = list(data.dtypes[data.dtypes == object].index)
num_feat = [f for f in data if f not in (cat_feat + ['Id'])]
cat_nunique = data[cat_feat].nunique()
cat_feat = list(cat_nunique[cat_nunique < 6].index)
dummy_data = pd.get_dummies(data[cat_feat], columns=cat_feat)
```

```python
dummy_data = pd.get_dummies(data[cat_feat], columns=cat_feat)
dummy_cols = list(set(dummy_data))
dummy_data = dummy_data[dummy_cols]
```

```python

new_data = pd.concat([data[num_feat],
                     dummy_data], axis=1)
```

```python
# Разделим выборку
X = new_data.loc[:, new_data.columns != 'SalePrice']
y = new_data['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
clf_rf = RandomForestRegressor(n_estimators=10, max_depth=5, min_samples_leaf=20, max_features=0.5, n_jobs=-1)
clf_rf.fit(X_train, y_train)   
```

```python
clf_rf.score(X_test, y_test)
```

```python
imp = pd.Series(clf_rf.feature_importances_, index = X_train.columns)
imp.sort_values(ascending=False)
```

3. Обучить стекинг как минимум 3х моделей, использовать хотя бы 1 линейную модель и 1 нелинейную

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
# scaler = StandardScaler()
```

```python
# scaler.fit_transform(X_train)
# scaler.fit_transform(X_test)
```

```python
regressor = StackingRegressor(
    [
        ('lr', LinearRegression()),
        ('knr', KNeighborsRegressor(n_neighbors = 7)),
    ],
RandomForestRegressor())
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred_proba_lr = regressor.named_estimators_['lr'].score(X_test, y_test)
y_pred_proba_dt = regressor.named_estimators_['knr'].score(X_test, y_test)
y_pred_proba = regressor.score(X_test, y_test)
```

```python
print(y_pred_proba_lr, y_pred_proba_dt, y_pred_proba)
```
