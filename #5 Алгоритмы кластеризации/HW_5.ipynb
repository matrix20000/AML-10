## 1. Визуально разделить город на районы безотносительно голосов (провести кластеризацию и вывести картинку)

```python
plt.scatter(X[:, 0], X[:, 1], color='Green', s=10)
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, AffinityPropagation
from sklearn.metrics import adjusted_rand_score, silhouette_score
```

#### Начнем с KMeans и подберем оптимальное количество кластеров

```python
k_inertia = []
ks = range(1,11)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(X_scaled, )
    k_inertia.append(clf_kmeans.inertia_)
plt.plot(ks, k_inertia)
plt.plot(ks, k_inertia ,'ro')
diff = np.diff(k_inertia)
plt.plot(ks[1:], diff)
```

```python
diff_r = diff[1:] / diff[:-1]
plt.plot(ks[1:-1], diff_r)
k_opt = ks[np.argmin(diff_r)+1]
k_opt
```

4 Района слишком мало, попробуем 8

```python
algorithm = KMeans(n_clusters=8)
clusters = algorithm.fit_predict(X_scaled)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='cool', s=15)
```

#### AgglomerativeClustering  и  AffinityPropagation сделать не могу, тк получаю ошибку MemoryError

#### DBSCAN()

```python
algorithm = DBSCAN()
clusters = algorithm.fit_predict(X_scaled)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='cool', s=15)
```
